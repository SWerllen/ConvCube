```python
import cv2

def find_interesting_points(image):
    """finds interesting points to track"""
    feature_params = dict( maxCorners = 100,
                           qualityLevel = 0.3,
                           minDistance = 7,
                           blockSize = 7 )
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    p0 = cv2.goodFeaturesToTrack(image, mask = None, **feature_params)
```
