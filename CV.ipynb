## Gather Images, Intersesting Points

```python
from convcube.cv.keypoints import interesting_points
from convcube.cv.keypoints import array2tuples
images = []
ipoints = []
corners = []
num_images = X_train_wy.shape[0]
for i in range(num_images):
    image = X_train_wy[i,:,:,:].transpose(1, 2, 0).copy()
    images.append(image)
    ipoints.append(interesting_points(image))
    corners.append(array2tuples(y_train_wy[i,:].reshape(4,2)))
```

## Get Close Interesting Points

```python
ix = 0
```

```python
from convcube.cv.keypoints import normalize_points

def get_close_points(corners, ips):
    """assumes both are normalized to 0,1"""
    ips = tuples2array(ips)
    pos, neg = [], []
    near_ix = np.zeros(ips.shape[0])
    far_ix = np.ones(ips.shape[0])
    for c in corners:
        c = np.array(c)
        dists = np.sqrt(np.sum((ips - c)**2, axis=1))
        near_ix[dists < 0.05] += 1
        far_ix[dists < 0.1] = 0
    near_ix = near_ix >= 1
    far_ix = far_ix > 0
    pos = array2tuples(ips[near_ix, :])
    neg = array2tuples(ips[far_ix, :])
    return pos, neg


img = images[ix].copy()
crns = corners[ix]
ips = interesting_points(img)
ips = normalize_points(ips, image)
pos, neg = get_close_points(crns, ips)
for pt in crns:
    x = int(pt[0] * 64.0)
    y = int(pt[1] * 64.0)
    img[y-1:y+1, x-1:x+1, :]  = (0, 255, 0)

for pt in pos:
    x = int(pt[0]*64.0)
    y = int(pt[1]*64.0)
    img[y-1:y+1, x-1:x+1, :]  = (255, 0, 0)
    
for pt in neg:
    x = int(pt[0] * 64.0)
    y = int(pt[1] * 64.0)
    img[y-1:y+1, x-1:x+1, :]  = (0, 0, 255)
    

plt.imshow(img[:,:,::-1])
ix += 1    
    
```

```python
from convcube.cv.keypoints import tuples2array, array2tuples
from convcube.cv.keypoints import normalize_points
from convcube.cv.keypoints import tuples2kpts
X, y = [], []
sift = cv2.SIFT(400)
for img, c, ip in zip(images, corners, ipoints):

    ip = normalize_points(ip, img)
    pos, neg = get_close_points(c, ip)
    pos, neg = tuples2kpts(pos, size=3), tuples2kpts(neg, size=3)
    
#     pos_X_1 = sift.compute(img[:,:,0], pos, None)[1]
#     pos_X_2 = sift.compute(img[:,:,1], pos, None)[1]
#     pos_X_3 = sift.compute(img[:,:,2], pos, None)[1]
#     pos_X = np.hstack([pos_X_1, pos_X_2, pos_X_3])
    pos_X = sift.compute(img, pos, None)[1]
#     neg_X_1 = sift.compute(img[:,:,0], neg, None)[1]
#     neg_X_2 = sift.compute(img[:,:,1], neg, None)[1]
#     neg_X_3 = sift.compute(img[:,:,2], neg, None)[1]
#     neg_X = np.hstack([neg_X_1, neg_X_2, neg_X_3])
    neg_X = sift.compute(img, neg, None)[1]

    
    if not pos_X is None:
        X.append(pos_X)
        y.append(np.ones(pos_X.shape[0]))
    if not neg_X is None:
        X.append(neg_X)
        y.append(np.zeros(neg_X.shape[0]))


X = [x for x in X if len(x.shape) == 2 and x.shape[1] == 128]
X = np.vstack(X)
y = np.concatenate(y)
print y.sum()
```

```python
from skimage.segmentation import slic
from skimage.segmentation import mark_boundaries
from skimage.segmentation import find_boundaries
def featurize_segments(image):
    """image -> segments"""
    segs = slic(image, n_segments=25, enforce_connectivity=True)

    features = []

    for i in range(segs.max()):
        masked = image[segs == i]
        
        #=====[ Pixel Values ]=====
        pix_avg = masked.mean(axis=0) #average pixel value
        pix_std = masked.std(axis=0) #std in pixel value
        pix_max = masked.max(axis=0)
        pix_min = masked.min(axis=0)
        pix_range = pix_max - pix_min

        #=====[ Moments ]=====
        Y, X = np.where(mask) 
        ylen = Y.max() - Y.min()
        xlen = X.max() - Y.max()
        yavg, ystd = Y.mean(), Y.std()
        xavg, xstd = X.mean(), X.std()
        area = len(X)
        
        #=====[ Put together ]====
        f = [ylen, xlen, yavg, ystd, xavg, xstd, area]
        f += pix_avg.tolist()
        f += pix_std.tolist()
        f += pix_max.tolist()
        f += pix_min.tolist()
        f += pix_range.tolist()
        features.append(f)
    
    return np.array(features)
```
