```python
Xs, ys = [], []
for frame in client.iter(Frame):
    if not frame['segment_features'] is None:
        X, y = frame['segment_features']
        if X.shape[1] == 22:
            Xs.append(X)
            ys.append(y)

X = np.vstack(Xs)
y = np.concatenate(ys)
print X.shape
```

## Fit Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import cross_val_score
clf = RandomForestClassifier()
clf.fit(X, y)
y_pred = clf.predict_proba(X)[:,1]
```

## Show Classification Pipeline

### i. original

```python
from convcube.convnets.pinpointing import get_X_pinpointing
from convcube.cv.segmentation import get_segs
from convcube.cv.segmentation import featurize_segs
frame = client.get_random(Frame)
image = get_X_pinpointing(frame['image'], frame['bounding_box'])[0,:,:,:].transpose(1, 2, 0)
plt.imshow(image[:,:,::-1])
```

### ii. segmented

```python
segs = get_segs(image)
plt.imshow(segs)
```

### iii. classified segments

```python
X = featurize_segs(image, segs)
y_pred = clf.predict_proba(X)[:,1] > 0.3
disp_img = image.copy()
for i in np.where(y_pred)[0]:
    disp_img[segs == i] = (0, 255, 0)
plt.imshow(disp_img[:,:,::-1])
```

### iv. segment centers

```python
disp_img = image.copy()
for i in np.where(y_pred > 0.3)[0]:
    mask = (segs == i)
    y, x = np.where(mask)
    yc, xc = (int(y.mean()), int(x.mean()))
    disp_img[yc-1:yc+1, xc-1:xc+1] = (0, 255, 0)
plt.imshow(disp_img[:,:,::-1])
```

## II. Hough lines bordering segments

### i. get edges from superpixel lines

```python
from skimage.segmentation import find_boundaries
boundaries = find_boundaries(segs.astype(np.uint8))
plt.imshow(boundaries, cmap=plt.cm.gray)
```

### ii. Hough Lines

```python
lines = cv2.HoughLines(boundaries.astype(np.uint8),1,np.pi/90,35)[0]
disp_img = image.copy()
for rho,theta in lines:
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 1000*(-b))
    y1 = int(y0 + 1000*(a))
    x2 = int(x0 - 1000*(-b))
    y2 = int(y0 - 1000*(a))

    cv2.line(disp_img,(x1,y1),(x2,y2),(0,0,255),2)
plt.imshow(disp_img[:,:,::-1])
```

```python
import numpy as np
from convcube.cv.segmentation import get_seg_centers
from ModalDB import ModalClient
from convcube.db import dbschema
from convcube.convnets.pinpointing import load_dataset_pinpointing

client = ModalClient('./data/db', schema=dbschema)
X_train, X_val, y_train, y_val = load_dataset_pinpointing(client, name='all')

print "Train shape: ", X_train.shape
print "Labels shape: ", y_train.shape
ix = 0
```

# Filter for only one labeled

```python
ix = (y_train != -1).sum(axis=1) == 8
X, Y = X_train[ix,:, :, :], y_train[ix, :]
ix = np.argmax(Y != -1, axis=1)
l = range(len(ix))
Y = np.array([Y[l,ix+i] for i in range(8)]).T
print 'Train data shape: ', X.shape
print 'Train labels shape: ', Y.shape
```

## Find projective transform

```python
from convcube.cv.keypoints import tuples2array, array2tuples
from convcube.cv.keypoints import normalize_points, denormalize_points
i = 10
s = 10
```

```python
x = X[i, :, :, :].transpose(1, 2, 0)
y = denormalize_points(array2tuples(Y[i,:].reshape(4, 2)), x)
```

```python
from skimage import transform as tf
y = tuples2array(y)
ix = np.argsort(y[:,0])
y = y[ix]

dst = np.array([(1*s, 1*s), (2*s, 1*s), (1*s, 2*s), (2*s, 2*s)])
tform = tf.ProjectiveTransform()
tform.estimate(dst, y)
```

```python
disp_img = x.copy()
for i in range(4):
    xc, yc = int(y[i,0]), int(y[i, 1])
    disp_img[yc, xc] = (0, 255, 0)
plt.imshow(disp_img[:,:,::-1])
```

```python
warped = tf.warp(disp_img, tform, output_shape=(3*s, 3*s))
plt.imshow(warped[:,:,::-1])
```
